```python
import pandas as pd
import utils
import json
```

```python
data_path = r'../Data/data_practicephase_cleardev/data_practicephase_cleardev/'
```

```python
data_list = []

for current_dataset in ['ArMIS','MD-Agreement','ConvAbuse', 'HS-Brexit']:                         # loop on datasets
  for current_split in ['train']:                                                                 # loop on splits, here only train 
    current_file = data_path+current_dataset+'_dataset/'+current_dataset+'_'+current_split+'.json'     # current file 
    data = json.load(open(current_file,'r', encoding = 'UTF-8'))                                  # load data 
    for item_id in data:                                                                          # loop across items for the loaded datasets                                                                                                  
      text = data[item_id]['text']        
      text = text.replace('\t',' ').replace('\n',' ').replace('\r',' ')                           # remove tabs and similar from text, so we can have everything on a line 
      data_list.append([current_dataset, current_split, item_id, data[item_id]['lang'], str(data[item_id]['hard_label']), str(data[item_id]['soft_label']["0"]), str(data[item_id]['soft_label']["1"]), str(data[item_id]['number of annotations']),  str(data[item_id]['annotations']), text])
df = pd.DataFrame(data_list, columns=['Dataset','Split','Id','Lang','Hard_label','Soft_label_0','Soft_label_1', 'NumAnnotations','Annotations', 'Text'])
```
